```python
import pandas as pd
a = pd.read_csv('./data/bert_data/spell_checker_result_Yiwise_data_3.txt')
Fasle_sentence = a[a['correct'] == False]
Fasle_sentence = Fasle_sentence[Fasle_sentence['init_word'] != '哎']
A = Fasle_sentence[Fasle_sentence['init_word'] != '喂']
right_texts = []
for i in range(len(A)):
    text = A.iloc[i]['text']
    init_word = A.iloc[i]['init_word']
    right_word = A.iloc[i]['max_word']
    right_text = text.replace(init_word,right_word)
    right_texts.append(right_text)
A['right_texts'] = right_texts
A.to_csv('/home1/shenxing/spell_checker_data.txt')
```

```python
A['score_diff'] = A['max_score'] - A['init_score']
```

```python
A1 = A[A['score_diff']>2]
A1 = A1[A1['max_score']>-10]
```
